```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data
data = {
    "Algorithm": [
        "K-Nearest Neighbor", "Logistic Regression", "Matrix Multiplication",
        "Array Sum", "Image Processing", "Quick Sort", "Merge Sort"
    ],
    "Serial Time (ms)": [11, 10, 8186, 7649, 81687, 20, 35],
    "Parallel Time (ms)": [5, 6, 1634, 1580, 4285, 37, 120]
}

# Create DataFrame
df = pd.DataFrame(data)

# Set style
sns.set_style("whitegrid")

# Creating line plot
plt.figure(figsize=(12, 6))
plt.plot(df["Algorithm"], df["Serial Time (ms)"], marker='o', label='Serial Time')
plt.plot(df["Algorithm"], df["Parallel Time (ms)"], marker='o', label='Parallel Time')
plt.title('Serial vs Parallel Time for Various Algorithms')
plt.xlabel('Algorithm')
plt.ylabel('Time (ms)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Creating bar plot
plt.figure(figsize=(12, 6))
bar_width = 0.35
index = range(len(df))

bar1 = plt.bar(index, df["Serial Time (ms)"], bar_width, label='Serial Time')
bar2 = plt.bar([p + bar_width for p in index], df["Parallel Time (ms)"], bar_width, label='Parallel Time')

plt.xlabel('Algorithm')
plt.ylabel('Time (ms)')
plt.title('Comparison of Serial and Parallel Times')
plt.xticks([p + bar_width / 2 for p in index], df["Algorithm"], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data
data = {
    "Algorithm": [
        "K-Nearest Neighbor", "Logistic Regression", "Matrix Multiplication",
        "Array Sum", "Image Processing", "Quick Sort", "Merge Sort"
    ],
    "Serial Time (ms)": [11, 10, 8186, 7649, 81687, 20, 35],
    "Parallel Time (ms)": [5, 6, 1634, 1580, 4285, 37, 120]
}

# Create DataFrame
df = pd.DataFrame(data)

# Set style
sns.set_style("whitegrid")

# Creating line plot with logarithmic scale
plt.figure(figsize=(12, 6))
plt.plot(df["Algorithm"], df["Serial Time (ms)"], marker='o', label='Serial Time')
plt.plot(df["Algorithm"], df["Parallel Time (ms)"], marker='o', label='Parallel Time')
plt.yscale('log')  # Set the y-axis to a logarithmic scale
plt.title('Serial vs Parallel Time for Various Algorithms (Log Scale)')
plt.xlabel('Algorithm')
plt.ylabel('Time (ms) (log scale)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Creating bar plot with logarithmic scale
plt.figure(figsize=(12, 6))
bar_width = 0.35
index = range(len(df))

bar1 = plt.bar(index, df["Serial Time (ms)"], bar_width, label='Serial Time')
bar2 = plt.bar([p + bar_width for p in index], df["Parallel Time (ms)"], bar_width, label='Parallel Time')

plt.xlabel('Algorithm')
plt.ylabel('Time (ms) (log scale)')
plt.title('Comparison of Serial and Parallel Times (Log Scale)')
plt.xticks([p + bar_width / 2 for p in index], df["Algorithm"], rotation=45)
plt.yscale('log')  # Set the y-axis to a logarithmic scale
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Algorithm": [
        "K-Nearest Neighbor", "Logistic Regression", "Matrix Multiplication",
        "Array Sum", "Image Processing", "Quick Sort", "Merge Sort"
    ],
    "Serial Time (ms)": [11, 10, 8186, 7649, 81687, 20, 35],
    "Parallel Time (ms)": [5, 6, 1634, 1580, 4285, 37, 120]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
for index, row in df.iterrows():
    plt.figure(figsize=(6, 4))
    bar_width = 0.35
    index = [0, 1]  # Positions for the bars

    # Plot bars for serial and parallel times
    plt.bar(index[0], row["Serial Time (ms)"], bar_width, label='Serial Time')
    plt.bar(index[1], row["Parallel Time (ms)"], bar_width, label='Parallel Time')

    plt.xlabel('Time Type')
    plt.ylabel('Time (ms)')
    plt.title(f'{row["Algorithm"]} Times Comparison')
    plt.xticks(index, ['Serial', 'Parallel'])
    plt.legend()
    plt.tight_layout()
    plt.show()
```
